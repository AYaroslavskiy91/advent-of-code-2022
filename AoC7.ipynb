```python
import numpy as np
import pandas as pd
```

```python
inp = open('aoc_input/aoc7.txt').read().rstrip()
commands = inp.split('$')
commands
```

```python
directories = {}
def get_dirs_and_files(command, current_directory):
    sub_dirs = []
    files = []
    for key in command.split('\n'):
        if 'dir' in key:
            sub_dirs.append(current_directory + '/' + key.split(' ')[-1])
        elif any(i.isdigit() for i in key):
            files.append(key)
    directory = {current_directory: sub_dirs + files}
    return directory
def parse_lines(commands):
    current_directory = '/'
    directories = {}
    for line in commands:
        command = line.split('\n')
        if 'cd ..' in command[0]:
            current_directory = get_parent(current_directory, directories)
        elif 'cd' in command[0] and '..' not in command[0]:
            current_directory = current_directory + '/' + command[0].split(' ')[-1]
        elif 'ls' in command[0]:
            directories.update(get_dirs_and_files(line, current_directory))
    return directories, current_directory
def get_directory_size(dir, directories):
    file_size = 0
    for item in dir:
        if any(i.isdigit() for i in item):
            file_size += (int(item.split(' ')[0]))
        else:
            file_size += (get_directory_size(directories[item], directories))
    return int(file_size)
def get_parent(child, directories):
    keys = [k for k, v in directories.items() if child in v]
    return keys[0]
```

```python
directories, cd = parse_lines(commands[2:])
sizes = {
    key: get_directory_size(directories[key],directories)
    for key in directories
}
```

```python
#part 1
solution = pd.Series([*sizes.values()])
solution[solution.le(100_000)].sum()
```

```python
#part 2
space_needed = sizes['/'] - 40_000_000
solution[solution.ge(space_needed)].min()
```
