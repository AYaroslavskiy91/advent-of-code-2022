```python
import numpy as np
import pandas as pd
```

```python
input = pd.read_csv('aoc_input/aoc4.txt', header=None).squeeze()
input.columns = ['first_section', 'second_section']
input
```

```python
first_input_parsed = [
    [int(row.split('-')[0]), int(row.split('-')[1])]
    for row 
    in input.first_section  
]
second_input_parsed = [
    [int(row.split('-')[0]), int(row.split('-')[1])]
    for row 
    in input.second_section  
]
```

```python
contained = 0
for index in range(len(input)):
    if (
        (first_input_parsed[index][0] <= second_input_parsed[index][1] and first_input_parsed[index][1] >= second_input_parsed[index][1])
        or (first_input_parsed[index][0] >= second_input_parsed[index][0] and first_input_parsed[index][1] <= second_input_parsed[index][1])
    ):
        contained +=1
contained
```

```python
overlap = 0
for index in range(len(input)):
    if (
        (first_input_parsed[index][1] < second_input_parsed[index][0])
        or (first_input_parsed[index][0] > second_input_parsed[index][1])
    ):
        continue
    else:
        overlap +=1
overlap
```
